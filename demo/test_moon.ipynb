```python
import plotly.express as px
import plotly as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons

from sklearn.cluster import DBSCAN
```

```python
sample_size = 500
x, y = make_moons(n_samples=sample_size,noise=0.05,random_state=5
                  )
```

```python
fig =px.scatter(x=x[:,0],y=x[:,1],color=y)
fig.show()
```

```python
db = DBSCAN(eps=0.2,min_samples=np.round(np.log(sample_size)))
db.fit(x)
```

```python
print(db.get_params())
```

```python
fig =px.scatter(x=x[:,0],y=x[:,1],color=db.labels_)
fig.show()
```

```python
time =  np.random.randint(0,50,sample_size)
np_st = np.stack((time,x[:,0],x[:,1]),1)
np_sorted = np_st[np.argsort(np_st[:,0])]
```

```python
fig =px.scatter( x = np_sorted[:,1],y=np_sorted[:,2],color=np_sorted[:,0])
fig.show()
```

```python
from src.st_dbscan import ST_DBSCAN

stdbscan = ST_DBSCAN(eps1=0.2,eps2=2,min_samples=np.round(np.log(sample_size)))
stdbscan.fit_frame_split(np_sorted,25)
```

```python
fig =px.scatter( x = np_sorted[:,1],y=np_sorted[:,2],color=stdbscan.labels, animation_frame=stdbscan.labels)
fig.update_layout(xaxis=dict(range=[-1.5,3]),
    yaxis=dict(range=[-1.5,1.5])
                  )
fig.write_html('fig_mon_split.html')
fig.show()
```

```python
stdbscanfit = ST_DBSCAN(eps1=0.2,eps2=2,min_samples=np.round(np.log(sample_size)))
stdbscanfit.fit(np_sorted)
```

```python
fig =px.scatter( x = np_sorted[:,1],y=np_sorted[:,2],color=stdbscanfit.labels, animation_frame=stdbscanfit.labels)
fig.update_layout(xaxis=dict(range=[-1.5,3]),
    yaxis=dict(range=[-1.5,1.5])
                  )
fig.write_html('fig_mon_fit.html')
fig.show()
```
